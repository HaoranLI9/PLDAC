```python
from IPython.display import SVG
from scipy import sparse
from sknetwork.visualization import svg_graph, svg_digraph, svg_bigraph

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
from IPython.core.display import Image, display

from matrice import matrice_adjacence, matrice_degree
from distance import distance_2graphe_diff_vertex_v2,distance_2graph
```

```python
img = Image(filename='french.jpg')
display(img)
```

```python
list_edge_french = [(0,1), (0,2), (0,4),(0,9), (0, 14),(3,4),(4,5), (4,6), (6,7), (7,8), (8,11), 
                    (10,11), (10, 12), (9,11),(9,13), (9,14)]
adjacency_french = matrice_adjacence(list_edge_french)
name_french = np.array(["Paris","Lille", "La Havre", "Brest", "La Mans", "Nantes", "Bordeaux", "Toulouse", "Montpellier", "Lyon", 
               "Marseille", "Avignon", "Nice", "Grenoble", "Dijon"] )
```

```python
adjacency = sparse.csr_matrix(adjacency_french)
image = svg_graph(adjacency, names=name_french)
SVG(image)
```

```python
degree_french = matrice_degree(adjacency_french)
index = np.argmax(degree_french.sum(axis = 1))
name_french[index]
```
